```python
import matplotlib
import numpy as np
```

```python
dir(matplotlib)
```

```python
from matplotlib import pyplot as plt
```

```python
dev_x=range(25,36)
dev_x[10]
```

```python
dev_y=[42105,44212,46324,51257,52541,55412,60124,62145,65555,68214,69125]
```

```python
plt.plot(dev_x,dev_y)
plt.show()
```

```python
plt.plot(dev_x, dev_y)

py_dev_y=dev_y+1000*(5-np.random.random(11))
plt.plot(dev_x, py_dev_y)

plt.xlabel("$ USD")
plt.ylabel("Ages(years)")
plt.title("Median Salary(USD) in USA")

# add legend to the plots in the order they were added or we can add label in plt.plot(label="label name") in each plot itself
plt.legend(["All dev", "python"])

plt.show()
```

```python
# add style to the plot 

# add dashed line style
plt.plot(dev_x, dev_y, 'k--', label="All dev") # or plt.plot(x,y,color='k', linestyle='--', label='All dev')

py_dev_y=dev_y+1000*(5-np.random.random(11))

# make the line blue
plt.plot(dev_x, py_dev_y, 'b', label="Python")

plt.xlabel("$ USD")
plt.ylabel("Ages(years)")
plt.title("Median Salary(USD) in USA")
plt.legend()

# add grid
plt.grid(True)
plt.show()
```
