```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t=np.arange(0.0,1.0,0.01)
s=np.sin(2*np.pi*t)
```

```python
# make line red
plt.rcParams['lines.color']='r'
plt.plot(t, s)
```

```python
c=np.cos(2*np.pi*t)
# make line thick
plt.rcParams['lines.linewidth']='3'
plt.plot(t, c)
plt.show()
```
