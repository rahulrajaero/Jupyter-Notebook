### Create ```movies_data``` database

````python
#Create TABLE
create_table = '''CREATE TABLE movies_data(
                    id BIGSERIAL PRIMARY KEY NOT NULL,
                    Title TEXT,
                    Directed_by TEXT,
                    Written_by TEXT,
                    Music_by TEXT,
                    Production_Company TEXT,
                    Distributed_by TEXT,
                    Running_Time TEXT,
                    Language TEXT,
                    Budget VARCHAR(100),
                    Box_Office VARCHAR(100));'''
cur.execute(create_table)
connection.commit()
print("Table created successfully in PostgreSQL ")
````

````python
select_query = "SELECT link FROM movies"
cur.execute(select_query)
url_data = cur.fetchall()
type(url_data)
````

````python
def get_content_value(row_data):
    return row_data.get_text(separator='\n', strip=True).splitlines()
````

#### Function to insert into ```movies_data``` table

````python
def insert_into_database(movie_info):
    movie_info
    
    title = movie_info['title']
    directed_by = movie_info.get('Directed by','None')
    written_by = movie_info.get('Written by','None')
    music_by = movie_info.get('Music by','None')
    production_company = movie_info.get('Production company','None')
    distributed_by = movie_info.get('Distributed  by','None')
    running_time = movie_info.get('Running time','None')
    language = movie_info.get('Language','None')
    budget = movie_info.get('Budget','None')
    box_office = movie_info.get('Box office','None')
    insert_query = '''INSERT INTO movies_data(title, directed_by, written_by, music_by, 
                        production_company, distributed_by, running_time,language, budget, box_office)
                        VALUES(%s, %s, %s,%s, %s,%s, %s,%s, %s, %s)'''


    record_to_insert = (title, directed_by, written_by, music_by, production_company[0], distributed_by[0], 
                        running_time[0],language, budget[0], box_office[0])
    cur.execute(insert_query, record_to_insert)
    connection.commit()
    print("Insertion Successful!")
````

````python
# get data from wikipedia
for url in url_data: # get url_data from 
        
    url = "https://en.wikipedia.org"+url[0]
    x = url.find('#')
    if x == -1:
        url = url
    else:
        url = url[:x]
    print(url)
    html = urlopen(url).read()
    soup = BeautifulSoup(html, 'html.parser')
    
    try:
        
        info_box = soup.find(class_="infobox vevent")
        info_rows = info_box.find_all('tr')

        # call the function to store the data
        movie_info = {}

        for index, row in enumerate(info_rows):
            if index == 0:
                movie_info['title'] = row.find("th").get_text(" ", strip=True)
            elif index == 1:
                continue
            else:
                content_key = row.find("th").get_text(" ", strip=True)
                content_value = get_content_value(row.find("td"))
                movie_info[content_key] = content_value

        # insert into databse
        insert_into_database(movie_info)
        
    except Exception as e:
        print(url)
        print(e)
````
