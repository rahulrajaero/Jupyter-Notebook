### Documents

````python
doc = '''This is first line. This lines describes nothing but natural language processing!
        A wood falls from the sky.'''
print(doc)
````

### Pre-Processing

    * Sentence Tokenizer
    * Word Tokenizer
    * Removing Punctuations
    * Remove StopWords
    * Stemming or Lemmatization
    

````python
import nltk
````

````python
# sentence Tokenization
sentences = nltk.sent_tokenize(doc)
print(type(sentences))
print(sentences)
````

````python
# words Tokenization
words = nltk.word_tokenize(doc)
print(type(words))
print(len(words))
print(words)
````

### Remove Punctuation

````python
no_punct_words = [word.lower() for word in words if word.isalpha()]
print(len(no_punct_words))
print(no_punct_words)
````

### Remove StopWords

````python
from nltk.corpus import stopwords

# list of stopwords
stopwords = stopwords.words('english')
#print(len(stopwords))
#print(stopwords)

# remove the stopwords
no_stop_words = [word for word in no_punct_words if word not in stopwords]
print(len(no_stop_words))
print(no_stop_words)
````

## Visualization
### WordCloud

````python
import numpy as np
from wordcloud import WordCloud
from PIL import Image
from matplotlib import pyplot as plt
````

````python
# use circle mask
mask = np.array(Image.open('circle.png')) # store circle.png in same folder

# generate wordcloud
wc = WordCloud(background_color='black', mask = mask).generate(doc)

# plot 
plt.figure(figsize = (10, 10))
plt.imshow(wc)
plt.axis('off')
plt.show()
````

### Stemming & Lemmatization
```Stemming``` truncate the words while ```Lemmatization``` tries to find the dictionary words.

````python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in no_stop_words]
print(stemmed_words)
````
